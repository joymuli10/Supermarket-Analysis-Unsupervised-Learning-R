# Data Cleaning

```
# Checking the number of unique values in each column
rapply(df,function(x)length(unique(x)))
```

```
# Checking for missing values
colSums(is.na(df))

# There are no null values
```

```
# Making all the columns lower cased for uniformity and replacing white spaces
names(df) <- str_replace_all(names(df), c(" " = "_"))
names(df) <- tolower(names(df))
```

```
# Checking for duplicated rows
duplicate_rows <- df[duplicated(df),]
duplicate_rows

# There are no duplicated rows
```

```
# Checking for outliers
num_cols <- unlist(lapply(df, is.numeric))
num_cols

# Subset numeric columns
data_num <- df[ ,num_cols]
```



> Outliers were in unit price, quantity, tax, cogs, gross margin percentage, gross income, rating & total
I will not deal with the outliers since in supermarkets total purchases vary.



# Exploratory Analysis

**Measures of central tendency**

```

# Getting the mean for each numeric column
colMeans(data_num)

# Getting the median for each numeric column
apply(data_num,2,median)
```

```
# Getting the mode
getmode <- function(v) {
   uniqv <- unique(v)
   uniqv[which.max(tabulate(match(v, uniqv)))]
}

getmode(df$unit.price)
getmode(df$quantity)
getmode(df$tax)
getmode(df$cogs)
getmode(df$gross.margin.percentage)
getmode(df$gross.income)
getmode(df$rating)
getmode(df$total)
```

```
# Histogram of various variables

hist(df$unit.price, col='cadetblue')
hist(df$quantity, col='cadetblue')
hist(df$tax, col='cadetblue')
hist(df$cogs, col='cadetblue')
hist(df$gross.margin.percentage, col='cadetblue')
#hist(df$grosss.income, col='cadetblue')
hist(df$rating, col='cadetblue')
hist(df$total, col='cadetblue')
```

# PCA

```
## Performing pca using prcomp()
df.pca <- prcomp(data_num[1:8], center = TRUE, scale. = FALSE)
df.pca
```

```
# The summary method of prcomp() also outputs the proportion of variance explained by the components.
summary(df.pca)

# PC1 explains 99.6% of the total variance
```

```
# Calling str() to have a look at your PCA object
str(df.pca)
```
